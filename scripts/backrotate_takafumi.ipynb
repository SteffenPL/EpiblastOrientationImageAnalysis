```python
import pandas as pd
import numpy as np
import re 
import os
from os.path import join, exists, basename, dirname
from glob import glob
import toml

def get_last_number(s):
    match = re.search(r'(\d+)$', s)
    if match:
        return int(match.group(1))
    else:
        raise ValueError(f"No ending number found in string: {s}")

pts_df = pd.read_csv(input_file)
pts_df["cluster_number"] = [get_last_number(cluster_id) for cluster_id in pts_df["cluster_id"]]
pts_df
```

```python
embryo_id_to_folder = {}
for folder in glob(join(data_dir, "epi*", "*")):
    image_id = basename(folder)
    embryo_id_to_folder[image_id] = folder

embryo_id_to_folder
```

```python
def backrotate_points(rotated_pts, R, offset, calibration, pixel_size, physical):
    """Rotate points into embryo coordinate frame."""
    factor = pixel_size[0] / calibration  # scaling between original voxel size and downscaled/rotated image voxel size

    pts = rotated_pts * factor * (1/calibration if physical else 1.0)
    pts = (R.T @ pts) + offset
    pts *= (calibration if physical else 1.0)
    return pts
```

```python
def rotate_points(pts, R, offset, calibration, pixel_size, physical):
    """Rotate points into rotated coordinate frame."""
    factor = pixel_size[0] / calibration  # scaling between original voxel size and downscaled/rotated image voxel size

    rotated_pts = (R @ (pts.T - offset)).T
    rotated_pts /= factor * (1/calibration if physical else 1.0)
    return rotated_pts
```

```python
# Coordinates are in order: ZYX
```

```python
# axis-0: Z
# axis-1: Y
# axis-2: X

for (index, row) in pts_df.iterrows():
    embryo_id = row["embryo_id"]

    # find image folder:
    image_folder = embryo_id_to_folder[image_id]

    rot_data_fn = join(image_folder, "snakemake", "rotated", "rotation_metadata.toml")
    rot_data = toml.load(rot_data_fn)

    R_image_to_embryo = np.array(rot_data["R_image_to_embryo"])
    offset = np.array(rot_data["offset"])
    calibration = 0.1632
    pixel_size = np.array(rot_data["pixel_size"])

    pts_img_coords = np.array([row["centroid_z"], row["centroid_y"], row["centroid_x"]])
    pts_phys_coords = np.array([row["centroid_z_um"], row["centroid_y_um"], row["centroid_x_um"]])
    
    pts_backrotated_img = rotate_points(pts_img_coords, R_image_to_embryo, offset, calibration, pixel_size, physical=False)
    pts_backrotated_phys = rotate_points(pts_phys_coords, R_image_to_embryo, offset, calibration, pixel_size, physical=True)

    pts_df.at[index, "centroid_x_rotated"] = pts_backrotated_img[2]
    pts_df.at[index, "centroid_y_rotated"] = pts_backrotated_img[1]
    pts_df.at[index, "centroid_z_rotated"] = pts_backrotated_img[0]

    pts_df.at[index, "centroid_x_um_rotated"] = pts_backrotated_phys[2]
    pts_df.at[index, "centroid_y_um_rotated"] = pts_backrotated_phys[1]
    pts_df.at[index, "centroid_z_um_rotated"] = pts_backrotated_phys[0]


#groupby embryo_id and write output files per embryo
for embryo_id, group in pts_df.groupby("embryo_id"):
    image_folder = embryo_id_to_folder[embryo_id]
    output_fn = join(image_folder, "snakemake", "rotated", f"rosette_cluster_centroids.csv")
    os.makedirs(dirname(output_fn), exist_ok=True)
    local_df = group.loc[:,["cluster_number", 
                            "centroid_x_rotated", 
                            "centroid_y_rotated", 
                            "centroid_z_rotated"]]
    # rename columns to index, axis-0, axis-1, axis-2
    local_df = local_df.rename(columns={
        "centroid_x_rotated": "axis-2",
        "centroid_y_rotated": "axis-1",
        "centroid_z_rotated": "axis-0",
        "cluster_number": "index"
    })
    # reorder columns
    local_df = local_df.loc[:,["index", "axis-0", "axis-1", "axis-2"]]
    local_df.to_csv(output_fn, index=False) 

    # also write original files (which were rotated)
    output_fn_orig = join(image_folder, "snakemake", "original", f"rosette_cluster_centroids.csv")
    os.makedirs(dirname(output_fn_orig), exist_ok=True)
    local_df_orig = group.loc[:,["cluster_number", "centroid_x", "centroid_y", "centroid_z"]]
    # rename columns to index, axis-0, axis-1, axis-2
    local_df_orig = local_df_orig.rename(columns={
        "centroid_x": "axis-2",
        "centroid_y": "axis-1",
        "centroid_z": "axis-0",
        "cluster_number": "index"
    })
    # reorder columns
    local_df_orig = local_df_orig.loc[:,["index", "axis-0", "axis-1", "axis-2"]]
    local_df_orig.to_csv(output_fn_orig, index=False)



# pts_df.to_csv(output_file, index=False)
```
